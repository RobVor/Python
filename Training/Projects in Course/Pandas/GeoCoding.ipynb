```python
import pandas, geopy

DataFrame4 = pandas.read_csv("http://pythonhow.com/supermarkets.csv", index_col="ID")
DataFrame4
```

```python
Nominating=geopy.geocoders.Nominatim(scheme="http")
```

```python
Location = Nominating.geocode("300 Kent Ave, Randburg, 2194")
Nominating.geocode("300 Kent Ave, Randburg, 2194")
```

```python
DataFrame4["GeoPoint"]=DataFrame4["Address"] + ", " + DataFrame4["City"] + ", " + DataFrame4["State"]
```

```python
DataFrame4
```

```python
DataFrame4["Latitude"]=DataFrame4["GeoPoint"].apply(Nominating.geocode)
```

```python
DataFrame4
```

```python
DataFrame4["Latitude"]=DataFrame4["Latitude"].apply(lambda x: x.latitude if x != None else None)
```

```python
DataFrame4
```

```python
DataFrame4["Longitude"]=DataFrame4["GeoPoint"].apply(Nominating.geocode)
DataFrame4["Longitude"]=DataFrame4["Longitude"].apply(lambda x: x.longitude if x != None else None)
```
