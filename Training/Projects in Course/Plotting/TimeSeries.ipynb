```python
from bokeh.plotting import figure, output_file, show
import pandas

DataFrame = pandas.read_csv("BTC USD Historical Data (2).csv", parse_dates=["Date"])

plot = figure(plot_width=1800,plot_height=900,x_axis_type="datetime")

plot.line(DataFrame["Date"], DataFrame["Price"], color="steelblue")

show(plot)
DataFrame
```
